## Question 2: Keys

__Briefly__ explain the following concepts:

- Primary Key
- Candidate Key
- Super Key
- Alternate Key
- Composite Key
- Unique Key
- Foreign Key

<u>Answer:</u>

- Primary Key: It is a column or a group of columns which uniquely identify all rows in the table (no two rows can exist with the same primary key). No value in this column can be NULL.
- Candidate Key: Any set of columns or group of columns which can also identify all rows in a table. One of the candidate keys is selected to be the primary key.
- Super Key: Any set of columns or group of columns which can also identify all rows in a table. A super key is a candidate key if it is minimal
- Alternate Key: It is the set of candidate keys which were not selected as the primary key
- Composite Key: It is a combination of two or more columns in a table that uniquely identify all rows in the table. Each individual columns need not be unique for every row.
- Unique Key: It can also be used to identify each row of a table. However, it allows one row to have a NULL (since no other row will have it, it will be unique)
- Foreign Key: A column or set of columns which are primary keys of another table, and maintain relationships between table.

## Question 5: More General Attribute Types

The relational model places restrictions on attributes. Many data scenarios have more complex types of attributes. __Briefly__ explain the following types of attributes:

- Simple attribute
- Composite attribute
- Derived attribute
- Single-value attribute
- Multi-value attribute



<u>Answer:</u>

- Simple attribute: An attribute that cannot be further subdivided (e.g. Firstname, ID number etc)
- Composite attribute: An attribute that can be split into further parts (e.g. Full name, Date of Birth, etc.)
- Derived attribute: An attribute that can be derived from other attributes (e.g. Total score, average marks etc.)
- Single-value attribute: An attribute that takes a single value for every entity (e.g. Age of a student)
- Multi-value attribute: An attribute that takes multiple values for every entity (contact details (mobile number, landline number, email etc.)

### Task 2: Complicated Queries

__Note:__ Performing the query in this task may require changing column values or types.

#### Query - Career Summary

- Write a query that produces a result of the form:
    - ```playerID```
    - ```nameLast```
    - ```nameFirst```
    - The sum of ```appearances.G_all``` for the player over all rows.
    - The sum over all rows of the following columns from batting:
        - ```G```
        - ```AB```
        - ```R```
        - ```AB```
        - ```2B```
        - ```3B```
        - ```HR```
        - ```RBI```
        - ```BB```
  
    - ```batting_average```, which is defined as $
\dfrac{sum(H)}{sum(AB)}
$

    - ```on_base_percentage```, which is defined as $
\dfrac{(sum(H) + sum(BB))}{(sum(AB) + sum(BB))}
$


- The query should be limited to 20 rows, and sorted by ```on_base_percentage``` from highest to lowest.


- ```batting_average``` and ```on_base_percentage``` should round to three decimal places.

````python
%%sql

with p1 as
    (select 
             playerID, 
             nameLast, 
             nameFirst 
        from people
    ),
    p2 as
    (select 
             playerID, 
             sum(G_all) as sum_G_all 
        from appearances 
        group by playerID
    ),
    p3 as
    (select 
            playerID, 
            sum(G) as sum_G, 
            sum(AB) as sum_AB, 
            sum(R) as sum_R, 
            sum(H) as sum_H, 
            sum(2B) as sum_2B, 
            sum(3B) as sum_3B, 
            sum(HR) as sum_HR,
            sum(RBI) as sum_RBI,
            sum(BB) as sum_BB 
        from batting
        group by playerID
    )
select 
        p1.playerID,
        p1.nameLast,
        p1.nameFirst,
        p2.sum_G_all,
        p3.sum_G,
        p3.sum_AB,
        p3.sum_R,
        p3.sum_H,
        p3.sum_2B,
        p3.sum_3B,
        p3.sum_HR,
        p3.sum_RBI,
        p3.sum_BB,
        p3.sum_H/p3.sum_AB as batting_average,
        (p3.sum_H+p3.sum_BB)/(p3.sum_AB+p3.sum_BB) as on_base_percentage
from
    (p1 join p2 using (playerID)) join p3 using (playerID)
order by on_base_percentage desc, nameLast asc
limit 20;
````

- Using a __subquery__, write a query that counts the number of people in the ```People``` table who do not have an entry in ```Appearances```.


- Run your query below. Note, your query will be __SLOW.__

````python
%%sql

select d.c4 - c.c3 as answer from
    (select count(distinct playerID) as c4 from people) as d, 
    (select count(a.c1) as c3
    from    (select distinct playerID as c1 from people) as a, 
            (select distinct playerID as c2 from appearances) as b
    where a.c1 = b.c2) as c;
````
